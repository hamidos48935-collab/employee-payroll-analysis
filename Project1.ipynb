# Salary distribution Analysis(Visualization)

```python
plt.figure()
sns.histplot(data['Salary'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
data['Salary'].mean()
```

```python
data['Salary'].median()
```

```python
plt.figure()
sns.boxplot(x=data['Salary'].dropna())
plt.title("Salary boxplot")
plt.show()
```

# Department-Level Analysis

```python
dept_counts = data['Department'].value_counts().head(10).transpose()
plt.figure()
dept_counts.plot(kind='bar')
plt.title('Top 10 Departments by Number of Employees')
plt.ylabel('Number of Employees')
plt.show()
```

```python
dept_salary = data.groupby('Department')['Salary'].mean().sort_values(ascending=False).head(10)
plt.figure()
dept_counts.plot(kind='bar')
plt.title('Top 10 Departments by Average Salary of Employees')
plt.ylabel('Number of Employees')
plt.show()
```

# Relationship Between  Typical Hours & Hourly Rate

```python
plt.figure()
sns.scatterplot(x=data['Typical Hours'].dropna(), y=data['Hourly Rate'])
plt.title('Typical Hours vs Hourly Rate')
plt.xlabel('Typical Hours')
plt.ylabel('Hourly Rate')
plt.show()
```

# Correlation Analysis

```python
numeric_data = data[['Typical Hours', 'Hourly Rate', 'Salary']]
plt.figure()
sns.heatmap(numeric_data.corr(), annot=True)
plt.title('Corrlation Matrix')
plt.show()
```
